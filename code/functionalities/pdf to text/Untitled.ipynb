```python
# img = cv2.imread("F:/major project/Functionalities/Text Recognition OCR/pdf to text/download.png")

import cv2
import matplotlib.pyplot as plt
img = cv2.imread(r"C:\Users\VAMC\Downloads\c70b8da4a33986f6d974ed15b788e1d5.jpg")
resized_img = cv2.resize(img, (664, 664), interpolation=cv2.INTER_NEAREST)
gray_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off') 
plt.subplot(122)
plt.imshow(gray_img, cmap='gray') 
plt.title("Preprocessed Image (Grayscale)")
plt.axis('off') 
plt.show()
```
